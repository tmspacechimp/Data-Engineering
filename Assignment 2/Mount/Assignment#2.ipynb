```python
info = pd.read_csv('characters_info.csv')
stats = pd.read_csv('character_stats.csv')
powers = pd.read_csv('superheroes_power_matrix.csv')
info_with_teams = info.assign(Team = [-1]*len(info['ID']))


ids = info_with_teams['ID'].tolist()


nteams = int(len(ids) / 5)
for i in range(nteams) :
    for j in range(5):
        id = randint(0, len(ids)-1)
        info_with_teams.loc[info_with_teams['ID'] == ids[id], 'Team'] = int(i)
        del ids[id]
```

```python
empty_strs = [''] * nteams
empty_ints = [0] * nteams

teams = pd.DataFrame({
    "team" : list(set(info_with_teams['Team'])),
    "team_name" : empty_strs,
    "leader" : empty_strs,
    "tank" : empty_strs,
    "damage" : empty_strs,
    "the_other_guy" : empty_strs,
    "the_other_guy2" : empty_strs,
    "top_average_speed" : empty_ints,
    "top_average_height" : empty_ints,
    "flight_count" : empty_ints
})

joint_table = info_with_teams.merge(stats, left_on = 'Name', right_on = 'Name', how = 'left')\
    .merge(powers, left_on = 'Name', right_on = 'Name', how = 'left')
```

```python
def get_team_members(team):
    return joint_table.loc[joint_table['Team'] == team]


def find_reamianing_candidates(team_members_df, picked):
    return team_members_df.loc[~joint_table['Name'].isin(picked)]


def find_team_name(team_members_df):
    hero_name = team_members_df.sort_values('Total', ascending = False)['Name'].values[0]
    return f"Team: {hero_name}"


def find_leader(team_members_df, picked):
    return  find_reamianing_candidates(team_members_df, picked).sort_values('Intelligence_x', ascending = False)\
        ['Name'].values[0]


def find_tank(team_members_df, picked):
    remaining =  find_reamianing_candidates(team_members_df, picked)
    tmp = pd.DataFrame({
        'Name' : remaining['Name'],
        'Sum' : remaining['Strength'] + remaining['Durability_x']
    })
    return tmp.sort_values('Sum', ascending = False)['Name'].values[0]


def find_damage(team_members_df, picked):
    remaining =  find_reamianing_candidates(team_members_df, picked)
    tmp = pd.DataFrame({
        'Name' : remaining['Name'],
        'Sum' : remaining['Speed'] + remaining['Power'] + remaining['Combat']
    })
    return tmp.sort_values('Sum', ascending = False)['Name'].values[0]


def find_other_guy(team_members_df, picked):
    return find_reamianing_candidates(team_members_df, picked)['Name'].values[0]


def find_top_avg_speed(team_members_df):
    return team_members_df.sort_values('Speed', ascending = False)['Speed'][0:3].mean()


def find_top_avg_height(team_members_df):
    return team_members_df.sort_values('Height', ascending = False)['Height'][0:3].mean()


def find_flight_count(team_members_df):
    return team_members_df.loc[team_members_df['Flight'] == True]['ID'].count()


def set_team_field(team, field, value):
    teams.loc[teams['team'] == team, field] = value

    
def set_fields(team):
    team_members = get_team_members(team)
    
    team_name = find_team_name(team_members)
    set_team_field(team, 'team_name', team_name)
    
    picked = []
    
    leader = find_leader(team_members, picked)
    set_team_field(team, 'leader', leader)
    picked.append(leader)
    
    tank = find_tank(team_members, picked)
    set_team_field(team, 'tank', tank)
    picked.append(tank)
    
    damage = find_damage(team_members, picked)
    set_team_field(team, 'damage', damage)
    picked.append(damage)
    
    the_other_guy = find_other_guy(team_members, picked)
    set_team_field(team, 'the_other_guy', the_other_guy)
    picked.append(the_other_guy)
    
    the_other_guy2 = find_other_guy(team_members, picked)
    set_team_field(team, 'the_other_guy2', the_other_guy2)
    
    top_average_speed = find_top_avg_speed(team_members)
    set_team_field(team, 'top_average_speed', top_average_speed)
   
    top_average_height = find_top_avg_height(team_members)
    set_team_field(team, 'top_average_height', top_average_height)
    
    flight_count = find_flight_count(team_members)
    set_team_field(team, 'flight_count', flight_count)
    
    
def fill_table():
    for i in list(teams['team']):
        set_fields(i)
        
```

```python
def get_team_row_from_name(team_name):
    return teams.loc[teams['team_name'] == team_name]


def get_top_avg_speed(team_data):
    return team_data['top_average_speed'].values[0]


def get_top_avg_height(team_data):
    return team_data['top_average_height'].values[0]


def get_flight_count(team_data):
    return team_data['flight_count'].values[0]


def get_peak_human_condition(team_data):
    id = team_data['team'].values[0]
    members = get_team_members(id)
    return members.loc[members['Peak Human Condition'] == True]['ID'].count()


def get_mind_control(team_data):
    id = team_data['team'].values[0]
    members = get_team_members(id)
    return members.loc[members['Mind Control'] == True]['ID'].count()


def get_good_guys(team_data):
    id = team_data['team'].values[0]
    members = get_team_members(id)
    return members.loc[members['Alignment'] == 'good']['ID'].count()


def face_off(team1, team2):
    team1_data = get_team_row_from_name(team1)
    team2_data = get_team_row_from_name(team2)
    pts1 = 0
    pts2 = 0
    
    speed_diff = get_top_avg_speed(team1_data) - get_top_avg_speed(team2_data)

    if speed_diff > 0:
        pts1 += 1
    elif speed_diff < 0:
        pts2 += 1
            
    height_diff = get_top_avg_height(team1_data) - get_top_avg_height(team2_data)

    if height_diff > 0:
        pts1 += 1
    elif height_diff < 0:
        pts2 += 1
            
    flight_diff = get_flight_count(team1_data) - get_flight_count(team2_data)

    if flight_diff > 0:
        pts1 += 1
    elif flight_diff < 0:
        pts2 += 1
                
    if pts1 == pts2: 
        phc_diff = get_peak_human_condition(team1_data) - get_peak_human_condition(team2_data) 
        if phc_diff > 0:
            return team1
        if phc_diff < 0:
            return team2
        
    return team2
```

```python
group_stage = pd.DataFrame({
    'team_name': teams['team_name'].values,
    'group': [-1] * nteams,
    'wins': [0] * nteams
})

def draw_groups():
    team_names = teams['team_name'].tolist()
    while not len(team_names) == 0:
        for i in range(4):
            if len(team_names) == 0:
                return
            index = randint(0, len(team_names) - 1)
            group_stage.loc[group_stage['team_name'] == team_names[index], 'group'] = i
            del team_names[index]

            
def add_win(winner):
    group_stage.loc[group_stage['team_name'] == winner, 'wins'] += 1
    
    
def simulate_group(group):
    team_names = group_stage.loc[group_stage['group'] == group, 'team_name'].tolist()
    for i in range(len(team_names) - 1):
        for j in range(i+1, len(team_names)):
            winner = face_off(team_names[i], team_names[j])
            add_win(winner)
            
            
def find_group_winner(group):
    teams = group_stage.loc[group_stage['group'] == group]
    max_wins = teams['wins'].max()
    pot_winners = teams.loc[group_stage['wins'] == max_wins]['team_name'].tolist()
    max_mind_control = 0
    winner = pot_winners[0]
    for team in pot_winners:
        mc = get_mind_control(get_team_row_from_name(team))
        if mc > max_mind_control:
            max_mind_control = mc
            winner = team
    return team
    
    
def simulate_groups():
    winners = []
    for i in range(4):
        simulate_group(i)
        winners.append(find_group_winner(i))
    return winners
```

```python
def simulate_semis():
    teams_and_wins = []
    for team in group_winners:
        teams_and_wins.append((team, group_stage\
            .loc[group_stage['team_name'] == team]['wins'].values[0]))   
    teams_and_wins.sort(key=lambda tup: tup[1], reverse=True) 
    winners = []
    winners.append(face_off(teams_and_wins[0][0], teams_and_wins[3][0]))
    winners.append(face_off(teams_and_wins[1][0], teams_and_wins[2][0]))
    return winners


finalists = simulate_semis()
finalists
```

- рЃЌрЃўрЃЌрЃЮрЃћрЃБрЃџрЃў рЃЊрЃдрЃўрЃА рЃЎрЃЮрЃюрЃфрЃћрЃарЃбрЃќрЃћ рЃАрЃфрЃћрЃюрЃљрЃќрЃћ рЃњрЃљрЃЏрЃЮрЃЊрЃўрЃА рЃќрЃБрЃАрЃбрЃљрЃЊ 7 рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃћрЃџрЃў
- рЃћрЃарЃЌрЃў рЃЊрЃљ рЃўрЃњрЃўрЃЋрЃћ рЃ»рЃњрЃБрЃцрЃерЃў рЃљрЃа рЃерЃћрЃўрЃФрЃџрЃћрЃЉрЃљ рЃўрЃДрЃЋрЃюрЃћрЃю рЃерЃћрЃЏрЃЊрЃћрЃњрЃў рЃЪрЃљрЃюрЃарЃћрЃЉрЃўрЃА рЃгрЃљрЃарЃЏрЃЮрЃЏрЃљрЃЊрЃњрЃћрЃюрЃџрЃћрЃЉрЃў:
    - k-pop рЃЊрЃљ rock/metal/punk/classic rock
    - thrash metal рЃЊрЃљ indie rock/rap
- рЃљрЃарЃф рЃћрЃарЃЌрЃў рЃљрЃа рЃБрЃюрЃЊрЃљ рЃўрЃДрЃЮрЃА рЃўрЃАрЃћрЃЌрЃў рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃћрЃџрЃў, рЃарЃЮрЃЏрЃџрЃўрЃА рЃАрЃљрЃ«рЃћрЃџрЃў рЃЕрЃЋрЃћрЃюрЃА рЃЉрЃљрЃќрЃљрЃерЃў рЃљрЃарЃАрЃћрЃЉрЃБрЃџ рЃАрЃ«рЃЋрЃљ рЃЏрЃЮрЃЏрЃдрЃћрЃарЃљрЃџрЃАрЃљрЃф/рЃ»рЃњрЃБрЃцрЃАрЃљрЃф рЃ░рЃЦрЃЋрЃўрЃљ. рЃћрЃарЃЌрЃў рЃЊрЃљ рЃўрЃњрЃўрЃЋрЃћ рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃџрЃљрЃЊ рЃЏрЃўрЃЋрЃўрЃЕрЃюрЃћрЃЋрЃЌ рЃЏрЃ«рЃЮрЃџрЃЮрЃЊ рЃЏрЃљрЃерЃўрЃю, рЃарЃЮрЃфрЃљ artist_name-рЃћрЃЉрЃў рЃЊрЃљ country-рЃћрЃЉрЃў рЃћрЃарЃЌрЃЏрЃљрЃюрЃћрЃЌрЃА рЃћрЃЏрЃЌрЃ«рЃЋрЃћрЃЋрЃљ
- рЃБрЃюрЃЊрЃљ рЃўрЃДрЃЮрЃА рЃћрЃарЃЌрЃў рЃЏрЃљрЃўрЃюрЃф рЃўрЃАрЃћрЃЌрЃў рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃћрЃџрЃў, рЃарЃЮрЃЏрЃћрЃџрЃЌрЃљ рЃАрЃўрЃЏрЃдрЃћрЃарЃћрЃЉрЃўрЃА (рЃарЃљрЃф songs.csv рЃцрЃљрЃўрЃџрЃерЃўрЃљ) рЃћрЃюрЃћрЃарЃњрЃўрЃБрЃџрЃЮрЃЉрЃўрЃА рЃ»рЃљрЃЏрЃў 100-рЃќрЃћ рЃЏрЃљрЃдрЃљрЃџрЃўрЃљ. рЃАрЃљрЃБрЃЉрЃљрЃарЃўрЃљ energy рЃЋрЃћрЃџрЃќрЃћ
- рЃ»рЃњрЃБрЃцрЃерЃў рЃерЃћрЃЏрЃљрЃЋрЃљрЃџрЃў рЃЌрЃўрЃЌрЃЮрЃћрЃБрЃџрЃў рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃџрЃўрЃА danceability рЃљрЃа рЃБрЃюрЃЊрЃљ рЃўрЃДрЃЮрЃА 30-рЃќрЃћ рЃюрЃљрЃЎрЃџрЃћрЃЉрЃў (рЃўрЃ«рЃўрЃџрЃћрЃЌ songs.csv рЃцрЃљрЃўрЃџрЃў). рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃџрЃўрЃА danceability рЃюрЃўрЃерЃюрЃљрЃЋрЃА рЃЏрЃўрЃАрЃў рЃАрЃўрЃЏрЃдрЃћрЃарЃћрЃЉрЃўрЃА danceability-рЃћрЃЉрЃўрЃА рЃ»рЃљрЃЏрЃА
- рЃЌрЃўрЃЌрЃЮрЃћрЃБрЃџ рЃ»рЃњрЃБрЃцрЃерЃў рЃљрЃБрЃфрЃўрЃџрЃћрЃЉрЃџрЃљрЃЊ рЃњрЃЋрЃДрЃљрЃЋрЃА рЃћрЃарЃЌрЃў рЃцрЃљрЃюрЃћрЃЉрЃўрЃА рЃАрЃљрЃДрЃЋрЃљрЃарЃћрЃџрЃў рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃћрЃџрЃў. рЃЏрЃўрЃА рЃЊрЃљрЃАрЃљрЃЊрЃњрЃћрЃюрЃљрЃЊ рЃБрЃюрЃЊрЃљ рЃЏрЃЮрЃЋрЃФрЃћрЃЉрЃюрЃЮрЃЌ top 7 fan-favourite рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃћрЃџрЃў, рЃарЃЮрЃЏрЃћрЃџрЃЌрЃљрЃњрЃљрЃю рЃЌрЃўрЃЌрЃЮрЃћрЃБрЃџрЃА рЃЕрЃЋрЃћрЃюрЃў 7 рЃ»рЃњрЃБрЃцрЃўрЃЊрЃљрЃю рЃћрЃарЃЌ-рЃћрЃарЃЌрЃерЃў рЃЕрЃљрЃЋрЃАрЃЋрЃљрЃЏрЃЌ (рЃЏрЃ«рЃЮрЃџрЃЮрЃЊ рЃћрЃарЃЌрЃерЃў). рЃЏрЃљрЃЌ рЃАрЃљрЃърЃЮрЃЋрЃюрЃћрЃџрЃљрЃЊ рЃБрЃюрЃЊрЃљ рЃЊрЃљрЃЋрЃЌрЃЋрЃљрЃџрЃЮрЃЌ рЃЋрЃћрЃџрЃў ***avg_scrobbles_per_listener***. рЃцрЃЮрЃарЃЏрЃБрЃџрЃљ: *scrobbles_lastfm / listeners_lastfm*. рЃћрЃА рЃЋрЃћрЃџрЃћрЃЉрЃў рЃљрЃарЃўрЃА top_artists.csv рЃцрЃљрЃўрЃџрЃерЃў. рЃњрЃљрЃўрЃЌрЃЋрЃљрЃџрЃўрЃАрЃгрЃўрЃюрЃћрЃЌ, рЃарЃЮрЃЏ fan-favourite рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃћрЃџрЃўрЃф рЃБрЃюрЃЊрЃљ рЃљрЃЎрЃЏрЃљрЃДрЃЮрЃцрЃўрЃџрЃћрЃЉрЃЊрЃћрЃА рЃќрЃћрЃЏрЃЮрЃЌ рЃљрЃдрЃгрЃћрЃарЃўрЃџ рЃЎрЃарЃўрЃбрЃћрЃарЃўрЃБрЃЏрЃћрЃЉрЃА
- 7 рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃџрЃўрЃАрЃњрЃљрЃю рЃерЃћрЃЏрЃЊрЃњрЃљрЃа рЃ»рЃњрЃБрЃцрЃА рЃАрЃљрЃ«рЃћрЃџрЃўрЃф рЃБрЃюрЃЊрЃљ рЃЊрЃљрЃљрЃарЃЦрЃЋрЃљрЃЌ top 7 fan-favourite-рЃА рЃЏрЃўрЃ«рЃћрЃЊрЃЋрЃўрЃЌ рЃерЃћрЃЏрЃЊрЃћрЃњрЃў рЃцрЃЮрЃарЃЏрЃљрЃбрЃўрЃЌ: 'Main Stage - рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃџрЃўрЃА рЃАрЃљрЃ«рЃћрЃџрЃў'

рЃарЃЮрЃЏ рЃерЃћрЃЋрЃљрЃ»рЃљрЃЏрЃЮрЃЌ, 7-рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃџрЃўрЃљрЃюрЃў рЃ»рЃњрЃБрЃцрЃўрЃЊрЃљрЃю рЃћрЃарЃЌ-рЃћрЃарЃЌрЃў рЃљрЃБрЃфрЃўрЃџрЃћрЃЉрЃџрЃљрЃЊ рЃљрЃарЃўрЃА рЃўрЃА, рЃарЃЮрЃЏрЃћрЃџрЃўрЃф top 7 fan-favourite-рЃерЃў рЃ«рЃЋрЃЊрЃћрЃЉрЃљ рЃЊрЃљ рЃЏрЃўрЃАрЃў рЃАрЃљрЃ«рЃћрЃџрЃў рЃћрЃарЃЦрЃЏрЃћрЃЋрЃљ рЃљрЃЏ рЃ»рЃњрЃБрЃцрЃАрЃљрЃф. рЃ«рЃЮрЃџрЃЮ рЃЊрЃљрЃюрЃљрЃарЃЕрЃћрЃюрЃў 6 рЃерЃћрЃЏрЃЌрЃ«рЃЋрЃћрЃЋрЃўрЃЌрЃЮрЃЉрЃўрЃА рЃърЃарЃўрЃюрЃфрЃўрЃърЃўрЃЌ рЃБрЃюрЃЊрЃљ рЃўрЃърЃЮрЃЋрЃЮрЃЌ. 7-рЃЋрЃћ рЃгрЃћрЃЋрЃарЃўрЃА рЃерЃћрЃЏрЃЌрЃ«рЃЋрЃћрЃЋрЃљрЃерЃў рЃЊрЃљрЃфрЃБрЃџрЃў рЃБрЃюрЃЊрЃљ рЃўрЃДрЃЮрЃА рЃДрЃЋрЃћрЃџрЃљ рЃќрЃћрЃЏрЃЮрЃЌрЃЕрЃљрЃЏрЃЮрЃЌрЃЋрЃџрЃўрЃџрЃў рЃърЃўрЃарЃЮрЃЉрЃљ.

**song list-рЃўрЃА рЃерЃћрЃарЃЕрЃћрЃЋрЃљ**

рЃ»рЃњрЃБрЃцрЃћрЃЉрЃўрЃА рЃЊрЃљрЃњрЃћрЃюрЃћрЃарЃўрЃарЃћрЃЉрЃўрЃА рЃерЃћрЃЏрЃЊрЃћрЃњ рЃЋрЃљрЃЊрЃњрЃћрЃюрЃЌ рЃЌрЃўрЃЌрЃЮрЃћрЃБрЃџрЃў рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃџрЃўрЃА рЃАрЃљрЃЎрЃЮрЃюрЃфрЃћрЃарЃбрЃЮ рЃърЃарЃЮрЃњрЃарЃљрЃЏрЃљрЃА. рЃЌрЃўрЃЌрЃЮрЃћрЃБрЃџрЃЏрЃљ рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃћрЃџрЃЏрЃљ рЃБрЃюрЃЊрЃљ рЃерЃћрЃљрЃАрЃарЃБрЃџрЃЮрЃА рЃЌрЃљрЃЋрЃўрЃАрЃў рЃбрЃЮрЃъ 3 рЃАрЃўрЃЏрЃдрЃћрЃарЃљ, рЃЮрЃдрЃЮрЃюрЃЊ рЃЊрЃљрЃфрЃБрЃџрЃў рЃБрЃюрЃЊрЃљ рЃўрЃДрЃЮрЃА рЃерЃћрЃЏрЃЊрЃћрЃњрЃў рЃЎрЃарЃўрЃбрЃћрЃарЃўрЃБрЃЏрЃћрЃЉрЃў:
- рЃАрЃўрЃЏрЃдрЃћрЃарЃљ рЃљрЃа рЃБрЃюрЃЊрЃљ рЃўрЃДрЃЮрЃА explicit (songs.csv рЃцрЃљрЃўрЃџрЃў)
- рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃћрЃџрЃЏрЃљ рЃБрЃюрЃЊрЃљ рЃўрЃЏрЃдрЃћрЃарЃЮрЃА рЃЌрЃљрЃЋрЃўрЃАрЃў рЃДрЃЋрЃћрЃџрЃљрЃќрЃћ рЃърЃЮрЃърЃБрЃџрЃљрЃарЃБрЃџрЃў рЃАрЃўрЃЏрЃдрЃћрЃарЃљ (popularity рЃЋрЃћрЃџрЃў рЃљрЃарЃўрЃА songs.csv-рЃерЃў)
- рЃБрЃюрЃЊрЃљ рЃўрЃЏрЃдрЃћрЃарЃЮрЃА рЃЌрЃљрЃЋрЃўрЃАрЃў рЃДрЃЋрЃћрЃџрЃљрЃќрЃћ рЃ«рЃЏрЃљрЃЏрЃљрЃдрЃљрЃџрЃў рЃАрЃўрЃЏрЃдрЃћрЃарЃљ (loudness). рЃарЃљрЃф рЃБрЃцрЃарЃЮ рЃЊрЃўрЃЊрЃўрЃљ рЃћрЃА рЃЋрЃћрЃџрЃў, рЃЏрЃўрЃЌ рЃБрЃцрЃарЃЮ рЃ«рЃЏрЃљрЃЏрЃљрЃдрЃљрЃџрЃўрЃљ рЃАрЃўрЃЏрЃдрЃћрЃарЃљ
- рЃБрЃюрЃЊрЃљ рЃўрЃЏрЃдрЃћрЃарЃЮрЃА  рЃЌрЃљрЃЋрЃўрЃАрЃў рЃарЃўрЃњрЃўрЃЌ рЃЏрЃћрЃЮрЃарЃћ рЃДрЃЋрЃћрЃџрЃљрЃќрЃћ рЃ«рЃљрЃюрЃњрЃарЃФрЃџрЃўрЃЋрЃў рЃАрЃўрЃЏрЃдрЃћрЃарЃљ.

рЃЉрЃЮрЃџрЃЮ рЃћрЃбрЃљрЃърЃўрЃљ рЃбрЃБрЃарЃюрЃћрЃА рЃЮрЃарЃњрЃљрЃюрЃўрЃќрЃћрЃЉрЃљ. рЃЌрЃЦрЃЋрЃћрЃю рЃЏрЃўрЃћрЃа рЃерЃћрЃарЃЕрЃћрЃБрЃџрЃЏрЃљ рЃ»рЃњрЃБрЃцрЃћрЃЉрЃЏрЃљ 1 рЃЎрЃЋрЃўрЃарЃўрЃА рЃњрЃљрЃюрЃЏрЃљрЃЋрЃџрЃЮрЃЉрЃљрЃерЃў рЃќрЃћрЃЏрЃЮрЃЌрЃ«рЃАрЃћрЃюрЃћрЃЉрЃБрЃџ рЃЦрЃљрЃџрЃљрЃЦрЃћрЃЉрЃерЃў рЃДрЃЋрЃћрЃџрЃњрЃљрЃю рЃБрЃюрЃЊрЃљ рЃЕрЃљрЃљрЃбрЃљрЃарЃЮрЃю рЃЎрЃЮрЃюрЃфрЃћрЃарЃбрЃћрЃЉрЃў. рЃфрЃ«рЃљрЃЊрЃўрЃљ, рЃћрЃарЃЌ рЃ»рЃњрЃБрЃцрЃА рЃћрЃарЃЌрЃарЃЮрЃБрЃџрЃљрЃЊ рЃАрЃ«рЃЋрЃљрЃЊрЃљрЃАрЃ«рЃЋрЃљ рЃЦрЃљрЃџрЃљрЃЦрЃерЃў рЃДрЃЮрЃцрЃюрЃљ рЃљрЃа рЃерЃћрЃБрЃФрЃџрЃўрЃљрЃЌ. рЃерЃћрЃАрЃљрЃЉрЃљрЃЏрЃўрЃАрЃљрЃЊ, рЃЏрЃАрЃњрЃљрЃЋрЃАрЃў scheduling рЃЎрЃЮрЃюрЃцрЃџрЃўрЃЦрЃбрЃћрЃЉрЃў рЃљрЃа рЃБрЃюрЃЊрЃљ рЃњрЃЦрЃЮрЃюрЃЊрЃћрЃЌ. рЃЊрЃўрЃюрЃљрЃЏрЃўрЃБрЃарЃљрЃЊ рЃЊрЃљ рЃерЃћрЃЏрЃЌрЃ«рЃЋрЃћрЃЋрЃўрЃЌрЃљрЃЊ рЃЊрЃљрЃљрЃњрЃћрЃюрЃћрЃарЃћрЃЌ рЃЏрЃљрЃбрЃарЃўрЃфрЃљ, рЃАрЃљрЃЊрЃљрЃф column-рЃћрЃЉрЃўрЃА рЃАрЃљрЃ«рЃћрЃџрЃћрЃЉрЃў рЃБрЃюрЃЊрЃљ рЃўрЃДрЃЮрЃА рЃцрЃћрЃАрЃбрЃўрЃЋрЃљрЃџрЃўрЃА рЃЏрЃўрЃЏрЃЊрЃўрЃюрЃљрЃарЃћрЃЮрЃЉрЃўрЃА рЃЊрЃдрЃћ (1-рЃЊрЃљрЃю 7-рЃЏрЃЊрЃћ), row-рЃћрЃЉрЃўрЃА рЃАрЃљрЃ«рЃћрЃџрЃћрЃЉрЃў (рЃљрЃюрЃБ рЃўрЃюрЃЊрЃћрЃЦрЃАрЃћрЃЉрЃўрЃА рЃАрЃљрЃ«рЃћрЃџрЃћрЃЉрЃў) рЃБрЃюрЃЊрЃљ рЃўрЃДрЃЮрЃА рЃЦрЃљрЃџрЃљрЃЦрЃўрЃА рЃАрЃљрЃ«рЃћрЃџрЃћрЃЉрЃў рЃЊрЃљ рЃЏрЃљрЃбрЃарЃўрЃфрЃљ рЃерЃћрЃЋрЃАрЃћрЃЉрЃБрЃџрЃў рЃБрЃюрЃЊрЃљ рЃўрЃДрЃЮрЃА рЃерЃћрЃЏрЃАрЃарЃБрЃџрЃћрЃЉрЃћрЃџрЃЌрЃљ рЃ»рЃњрЃБрЃцрЃћрЃЉрЃўрЃА рЃАрЃљрЃ«рЃћрЃџрЃћрЃЉрЃўрЃЌ.
